News Category Dataset

Η εργασία αφορά ένα σύνολο δεδομένων που περιέχει 200.000 τίτλους ειδήσεων κατηγοριοποιημένα βάση τίτλου, κατηγορίας, σύντομης περιγραφής, συγγραφέα και ημερομηνίας. Σκοπός της εργασίας είναι η αναγνώριση του τύπου της είδησης με βάση τον τίτλο και την σύντομη περιγραφή της με τη βοήθεια
ενός τεχνητού νευρωνικού δικτύου δηλαδή ενός μοντέλου που θα μπορεί να προβλέψει μια ακολουθία λέξεων με βάση την είσοδο που του δόθηκε. Το μοντέλο που θα χρησιμοποιήσουμε σε αυτή τη περίπτωση είναι το RNN (Recurrent Neural Network). Πιο συγκεκριμένα το RNN μπορεί να απομνημονεύει 
προηγούμενες εισόδους και χρησιμοποιείται ειδικότερα στη ταξινόμηση αρχείων, λέξεων και για αυτό το λόγο είναι και το πιο κατάλληλο μοντέλο 
Αποτελείται από τρία μέρη: την είσοδο(input), εάν κρυφό επίπεδο(hidden layer) και ένα αποτέλεσμα(output) όπως φαίνεται και στη παρακάτω εικόα.

Ο τρόπος που λειτουργεί το RNN είναι ο εξής. H είσοδος λαμβάνει τις πληροφορίες για επεξεργασία και το επίπεδο εξόδου παρέχει το αποτέλεσμα. Η επεξεργασία, η ανάλυση και η πρόβλεψη δεδομένων πραγματοποιούνται στο κρυφό επίπεδο. To hidden layer είναι χρήσιμο διότι απομνημονεύει τις προηγούμενες εισόδους και αναλαμβάνει ενεργό ρόλο στην κατάληξη των επομένων αποτελεσμάτων. Όπως φαίνεται και στην εικόνα κάθε είσοδος θα πάρει υπόψιν και τις προηγούμενες εισόδους με αποτέλεσμα η κάθε έξοδος να είναι πιο κοντά στο να μπορεί να κάνει καλύτερες προβλέψεις.

Όμως υπάρχουν δυο κύρια προβλήματα που αντιμετωπίζουμε σχετικά με το συγκεκριμένο μοντέλο. Το πρώτο ονομάζεται vanishing gradient problem και το δεύτερο gradient exploding problem. Το gradient είναι ένα διάνυσμα παραγώγων μιας συνάρτησης που σχετίζεται με τις εισόδους και μεταφέρεται σε κάθε επόμενη κατάσταση(state). Ο σκοπός του είναι να αντιπροσωπεύει πόσο σημαντική είναι η κάθε είσοδος στην τελική έξοδο.

Τα συγκεκριμένα προβλήματα συμβαίνουν όταν η ακολουθία είναι πολύ μεγάλη. Δηλαδή το μοντέλο "δυσκολεύεται" να διατηρήσει πληροφορίες ακολουθιών που είναι πολύ μεγάλες με αποτέλεσμα σε κάθε επιπλέον κατάσταση να μην διατηρούνται όλες οι πληροφορίες. Στη πρώτη περίπτωση(vanishing gradient problem) το gradient αποκτά πολύ μικρές τιμές που κοντεύουν στο 0 με συνέπεια να χαθούν οι πληροφορίες όσο πιο πολύ προχωράει ο αλγόριθμος και το μοντέλο να μην μπορεί να μάθει βάση των προηγουμένων ακολουθιών. Στη δεύτερη περίπτωση που μας απασχολεί και σε αυτή την εργασία (gradient exploding problem) το gradient παίρνει πολύ μεγάλες τιμές με αποτέλεσμα όσο πιο πολύ προχωράει ο αλγόριθμος τόσο μεγαλύτερες τιμές παίρνει, μέχρι κάποια στιγμή να "κρασαρει".

Μια λύση γι αυτό το πρόβλημα είναι η αρχιτεκτονική LSTM(Long Short-Term Memory) η οποία επεξεργάζεται ολόκληρες ακολουθίες δεδομένων με τη βοήθεια ανατροφοδότησης των προηγουμένων εισόδων. 
Η αρχιτεκτονική του LSTM αποτελείται από τρία κύρια μέρη:
    Forget gate: Αποφασίζει ποιες πληροφορίες θα κρατηθούν και ποιες όχι. 
    Input gate: Αφού σταλθούν οι πληροφορίες που θα κρατηθούν αυτό το στάδιο αποφασίζει το μέγεθος που θα πρέπει να μεταφέρει στο επόμενο στάδιο
    Output gate: Είναι υπεύθυνο για την εμφάνιση αποτελεσμάτων καθώς και τη μεταφορά των πληροφοριών στο κρυφό επίπεδο για την έναρξη νέας κατάσταης.

Παρακατω φαινεται ο κωδικας του συγκεκριμενου μοντελου:

```python
#libraries we need to import
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense, Concatenate
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#load the dataset
data = pd.read_json(r"C:\Users\eleni\Documents\News_Category_Dataset_v3.json", lines=True)
#display the first 100 rows
data.head(100)
```

```python
#show all the categories and how many each are
data["category"].value_counts()
```

Αφού φορτώσουμε τα δεδομένα μας, εξάγουμε μόνο τις πληροφορίες που χρειαζόμαστε και τις αποθηκεύουμε σε μεταβλητές. Σε αυτή τη περίπτωση headlines.descriptions,categories.

```python
#extract the data from these 3 specific datasets
headlines = data['headline'].values
descriptions = data['short_description'].values
categories = data['category'].values
```

Στη συνέχεια με τη βοήθεια του tokenizer κάθε λέξη μετατρέπεται σε μια ακολουθία(token) και στη κάθε ακολουθία αναθέτεται ένας ξεχωριστός αριθμός. Εδώ έχουμε ορίσει ο μέγιστος αριθμός λέξεων να είναι 10000 (οι λέξεις που μετατρέπει είναι οι πιο συχνές λέξεις).Αποθηκεύουμε τις ακολουθίες σε δυο διαφορετικές μεταβλητές και τις κάνουμε padding για να έχουν όλες το ίδιο μέγεθος.

```python
#converts text into a sequence of integers with Tokenizer
tokenizer = Tokenizer(num_words=10000) #keep top 10000 most frequent words
tokenizer.fit_on_texts(headlines + descriptions) #assigns a unique ID integer for each unique word
head_sequences = tokenizer.texts_to_sequences(headlines)
desc_sequences = tokenizer.texts_to_sequences(descriptions)
```

```python
#pad the sequences (length of 100 in this case)
max_length = 100 
X_head = pad_sequences(head_sequences, maxlen=max_length)
X_desc = pad_sequences(desc_sequences, maxlen=max_length)
```

Η κατηγορία(category) χρησιμοποιεί ένα διαφορετικό tokenizer και αποθηκεύεται σε διαφορετική μεταβλητή.

```python
#convert the category wirh a different tokenizer
category_tokenizer = Tokenizer()
category_tokenizer.fit_on_texts(categories)
category_sequences = category_tokenizer.texts_to_sequences(categories)
y = np.array([seq[0] - 1 for seq in category_sequences]) #starts from the first element of the sequence
```

Mε τη βοήθεια της βιβλιοθήκης sklearn.model_selection χρησιμοποιούμε τα function train-test για να εκπαιδεύσουμε τα δεδομένα μας. Σε αυτή τη περίπτωση χρησιμοποιούμε το 80% των δεδομένων για εκπαίδευση και το 20% δεδομένων για testing. Τα δεδομένα που χρησιμοποιούνται για εκπαίδευση βοηθάνε στη δημιουργία του μοντέλου και στη συνέχεια συγκρίνονται με το υπόλοιπο ποσοστό για να βρεθούν τυχόν σφάλματα που εμφανίστηκαν και να προσαρμοστεί το μοντέλο καταλλήλως

```python
#train-test split, in this case 20% is used for testing while the remaining 80% is used for training
X_train_head, X_test_head, X_train_desc, X_test_desc, y_train, y_test = train_test_split(X_head, X_desc, y, test_size=0.2, random_state=0)
```

```python
#define model parameters
vocab_size = 10000
embedding_dim = 150
num_categories = len(category_tokenizer.word_index)
```

Στο τελευταίο στάδιο περνάει από ένα επίπεδο embedding, στο οποίο η κάθε ακολουθία μπαίνει σε ένα διάνυσμα συγκεκριμένων διαστάσεων (σε αυτή τη περίπτωση 150) και μετά περνάει στο μοντέλο lstm το οποίο αναλύει το συγκεκριμένο διάνυσμα με το τρόπο που αναφέρθηκε παραπάνω. Τα συγκεκριμένα αποτέλεσμα του headline και του description με τη βοήθεια του concatenate ενώνονται μεταξύ τους, περνάνε σε ένα διάνυσμα και με το reLU οι αρνητικές τιμές μετατρέπονται σε 0 έτσι ώστε το μοντέλο να επικεντρωθεί στις τιμές που είναι πιο σημαντικές. Τέλος με τη βοήθεια του softmax το διάνυσμα της κατηγορίας μετατρέπεται σε πιθανότητα και αντιστοιχίζεται με το διάνυσμα του τίτλου/περιγραφής προβλέποντας έτσι σε ποια κατηγορία ανήκει ο τίτλος της κάθε είδησης.

```python
input_head = Input(shape=(max_length,))
embedding_head = Embedding(input_dim=vocab_size, output_dim=embedding_dim)(input_head)
lstm_head = LSTM(150, return_sequences=False)(embedding_head)

input_desc = Input(shape=(max_length,))
embedding_desc = Embedding(input_dim=vocab_size, output_dim=embedding_dim)(input_desc)
lstm_desc = LSTM(150, return_sequences=False)(embedding_desc)

concat = Concatenate()([lstm_head, lstm_desc])
dense = Dense(64, activation='relu')(concat)
output = Dense(num_categories, activation='softmax')(dense)

model = Model(inputs=[input_head, input_desc], outputs=output)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#train the model
history = model.fit([X_train_head, X_train_desc], y_train,epochs=10,batch_size=40,validation_data=([X_test_head, X_test_desc], y_test))
```

Παραπάνω βλέπουμε κάποια αποτελέσματα που βρέθηκαν σε κάθε προσπάθεια. Το accuracy αντιπροσωπεύει τον αριθμό των σωστών προβλέψεων που έχουν γίνει από το μοντέλο, ενώ το val_accuracy αντιπροσωπεύει τον αριθμό των σωστόν προβλέψεων που έγιναν από το μοντέλο πάνω στα δεδομένα που χρησιμοποιούνται για testing. Αντίστοιχα το loss είναι ο αριθμός των σφαλμάτων που έκανε το μοντέλο. Όσο μικρότερος είναι τόσο πιο επιτυχημένο είναι το μοντέλο, αντίστοιχα και το val_loss αντιπροσωπεύει τα στοιχεια που χρησιμοποιήθηκαν για testing.

Στη συνέχεια ακολουθεί το μοτίβο που θα ακολουθήσει το μοντέλο στις επόμενες ακολουθίες αφού έχει εκπαιδευτεί. Όπως βλέπουμε ακολουθεί τις ίδιες διαδικασίες με πριν(χρήση tokenizer και padding) με τη μόνη διαφορά να είναι ότι το μοντέλο θα χρησιμοποιηθεί αυτόματα χωρίς να επαναλάβουμε οδηγίες. Πιο συγκεκριμένα θα πάρει τις ακολουθίες των headline,description και θα φτιάξει ένα πίνακα που κάθε γραμμή θα αντιπροσωπεύει τις πιθανότητες για κάθε ακολουθία και κάθε στήλη αντιπροσωπεύει τη κατηγορία που βρίσκεται σε μορφή αριθμού. Με το argmax διαλέγει τη μεγαλύτερη πιθανότητα ενώ έχουμε διαλέξει να ελέγχει μόνο κάθε γραμμή. Τέλος αντιστοιχείται ο αριθμός με την μεγαλύτερη πιθανότητα με την κατηγορία.

```python
#function to predict the next sequence
def predict_category(headline, description):
    head_sequence = tokenizer.texts_to_sequences([headline])
    desc_sequence = tokenizer.texts_to_sequences([description])

    head_padded = pad_sequences(head_sequence, maxlen=max_length)
    desc_padded = pad_sequences(desc_sequence, maxlen=max_length)
    
    prediction = model.predict([head_padded, desc_padded])
    category_idx = np.argmax(prediction, axis=1)[0] # finds the index of the highest probability in the prediction array

    predicted_category = list(category_tokenizer.word_index.keys())[list(category_tokenizer.word_index.values()).index(category_idx + 1)]
    
    return predicted_category
```

```python
#example
input_headline = 'Home Team Wins Championship After Thrilling Overtime Finish'
input_description = 'The local sports team clinched the championship title after an intense and exciting overtime match, bringing joy to their fans.'
predicted_category = predict_category(input_headline, input_description)
print(predicted_category)
```

```python
#multiple examples
examples = [
    ("New Study Reveals the Benefits of Daily Exercise", "Researchers have discovered that incorporating a daily exercise routine can significantly improve mental and physical health, leading to longer life expectancy."),
    ("Tech Giant Releases Latest Smartphone Model", "The newest smartphone from the tech giant features groundbreaking technology, including an advanced camera system and enhanced battery life."),
    ("Local Bakery Wins National Award for Best Bread", "A small bakery in the heart of the city has been recognized nationally for its outstanding bread, drawing attention from food enthusiasts across the country."),
    ("Government Announces New Environmental Policies", "The government has introduced new policies aimed at reducing carbon emissions and promoting sustainable practices to combat climate change."),
    ("Sports Team Makes Historic Comeback in Championship", "In an incredible turn of events, the underdog team secured a victory in the championship game after being behind by 20 points, making history in the league."),
    ("Innovative Startup Launches Eco-Friendly Product Line", "A forward-thinking startup has launched a new line of eco-friendly products designed to reduce environmental impact and promote sustainability."),
    ("Famous Artist Exhibits New Collection in Major Gallery", "The renowned artist has unveiled a new collection of paintings at a major gallery, drawing large crowds and critical acclaim."),
    ("Scientists Discover Potential Cure for Rare Disease", "A team of scientists has made a breakthrough in the treatment of a rare disease, offering hope to thousands of affected patients."),
    ("City Council Approves Plans for New Park", "The city council has approved plans to develop a new park in the downtown area, providing a green space for residents to enjoy."),
    ("Record-Breaking Heatwave Sweeps Across the Country", "A historic heatwave is affecting millions of people across the country, with temperatures reaching all-time highs and causing widespread disruption.")
]

for headline, description in examples:
    predicted_category = predict_category(headline, description)
    print(f"Headline: {headline}\nDescription: {description}\nPredicted Category: {predicted_category}\n")
```
